```python
import folium

# Calculate the top 50 most frequent stops
top_50_stops = df['stop_name'].value_counts().nlargest(50).index

# Filter the dataframe to include only the top 50 stops
top_50_df = df[df['stop_name'].isin(top_50_stops)]

# Create a map centered around the average location of the top 50 stops
map_center_lat = top_50_df['stop_lat'].mean()
map_center_lon = top_50_df['stop_lon'].mean()

m = folium.Map(location=[map_center_lat, map_center_lon], zoom_start=10)

# Add markers for each of the top 50 stops
for index, row in top_50_df.iterrows():
    folium.Marker([row['stop_lat'], row['stop_lon']], popup=row['stop_name']).add_to(m)

# Display the map
m
```
